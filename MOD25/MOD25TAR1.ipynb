1. Marque quais desses métodos/algoritmos muito populares em ciência de dados são baseados no uso de derivada:

    1. Método Mínimos Quadrados
    2. Gradiente descendente
    3. Newton Raphson
    4. CART (Árvore de decisão)

| Método | Uso de Derivada | Explicação |
| :--- | :---: | :--- |
| **Mínimos Quadrados** | **Sim** | Para minimizar o erro quadrático em uma regressão linear, derivamos a função de custo em relação aos coeficientes e igualamos a zero para encontrar o ponto de mínimo. |
| **Gradiente Descendente** | **Sim** | É a base das redes neurais e do GBM. Utiliza o vetor gradiente (derivadas parciais) para apontar a direção de descida na curva de erro. |
| **Newton-Raphson** | **Sim** | Um método de otimização de segunda ordem que utiliza a primeira e a segunda derivada (Hessiana) para encontrar raízes de funções de forma muito rápida. |
| **CART (Árvore)** | **Não** | Árvores de decisão utilizam critérios estatísticos de partição (como Índice Gini ou Entropia) e busca exaustiva (Greedy Search), não dependendo de funções contínuas ou deriváveis para realizar os cortes. |

### Conclusão:
A maioria dos modelos de aprendizado estatístico que tratam a otimização como um problema **contínuo** (onde podemos "deslizar" por uma curva de erro) utiliza derivadas. Já os modelos que tratam o problema de forma **discreta** ou por "quebras" nos dados (como as Árvores de Decisão) seguem uma lógica de busca lógica e comparativa.

2. Dada uma base de dados com uma variável resposta $y$ e um conjunto de variáveis explicativas. Considere uma estrutura de um modelo de regressão. Explique com suas palavras por que não é possível obter parâmetros que forneçam um erro quadrático médio (EQM) menor que o obtido com estimadores de mínimos quadrados.

O Método de Mínimos Quadrados fornece o menor Erro Quadrático Médio (EQM) possível para um modelo linear devido a dois fatores principais:

1. **Otimização via Cálculo:** A função de erro (EQM) em modelos lineares é **convexa**, possuindo o formato de uma "tigela". O método de Mínimos Quadrados utiliza a derivada para encontrar o ponto exato de **mínimo global** (o fundo da tigela). Matematicamente, qualquer outro parâmetro escolhido estará em um ponto mais alto da curva, resultando em um erro maior.

2. **Projeção Geométrica:** Geometricamente, o menor erro possível é a distância mais curta entre os dados reais e a reta de regressão. Essa distância mínima é sempre a **projeção ortogonal** (perpendicular). O estimador de Mínimos Quadrados é justamente o cálculo dessa projeção; qualquer inclinação diferente aumentaria a distância dos resíduos.



**Conclusão:** Como o OLS encontra o ponto onde a inclinação da função de erro é zero, ele atinge o limite matemático inferior do erro para o conjunto de dados de treino.
